```
df = pd.read_csv(Config.example_ground_truth_path)
features_df = pd.read_csv(Config.features_path)



# Combine variables we want in one place
df = df.drop(['image_id','seborrheic_keratosis'],axis=1)
df['area'] = features_df['area']
df['perimeter'] = features_df['perimeter']
df['ratio'] = df['area'] / df['perimeter']

colors1 = ['#8BA888','#D67573']


colors = ["green" if df['melanoma'][x] == 0.0 else 'red' for x in range(len(df['melanoma'])) ]
print(df.head())
sns.scatterplot(df['area'],df['perimeter'], hue= colors, palette= colors1)

```

```
plt.scatter(df['area'],np.log(df['perimeter']+1), c=colors, edgecolors='white')
```
